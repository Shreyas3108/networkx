```python
g = networkx.read_edgelist("facebook_combined.txt",create_using=networkx.Graph(),nodetype=int)
```

```python
print networkx.info(g)
```

```python
sp = networkx.spring_layout(g)
```

```python
plt.title('Facebook Data')
plt.axis('Off')
networkx.draw_networkx(g,pos=sp,with_labels = False,node_size = 20,node_color='blue')
plt.show()
```
